#### __Variable:__
A variable is a labeled container that is used to store value(s). Look at it as a box where you label it with a name and store your item(s) in it.

**What is the purpose of a variable?**

It is to store, reference, and manipulate information.

**Naming Variables:**
* A variable name must start with a letter or the underscore character
* A variable name cannot start with a number
* A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
* Variable names are case-sensitive (number, Number, and NUMBER are three different variables)
* A variable name cannot be any of the Python keywords.

````python
# >> Instructions: Follow the comments step-by-step
# We will create a variable named 'score' and assign it a value of '90'. Note: to assign a value we use '='

score = 90
# To display a variable we use the 'print' function. This is an example of referencing a variable
# print the score variable
print(score)

# To manipulate a variable, Add 1 to the score value and update it. 
# Score should be 91.

score = score+1
print(score)

````

````python
# We have a list of fruits, let's print each fruit in the list separately.
fruits = ["banana", "orange", "apple"]

````

````python
# Let's change 'banana' to 'kiwi' & 'orange' to 'strawberry'

fruits[0] = 'kiwi'
fruits[1] = 'strawberry'

# The fruits list should now looks like ["kiwi", "strawberry", "apple"]
fruits
````

#### List methods

````python
# We have a list of fruits to buy
fruits = ["orange"]
# We need to add 'kiwi' to our list. Using append()

fruits.append('kiwi')
# fruits is now: ["orange", "kiwi"]
print(fruits)
````

````python
# Maybe we should get 'lemon' before 'kiwi' . Using insert()

fruits.insert(1, 'lemon')
# fruits is now: ["orange", "lemon", "kiwi"]
print(fruits)
````

````python
# Remove the 'kiwi' from our list. Using pop()

fruits.pop()
# fruits is now: ["orange", "lemon"]
print(fruits)

# Remove 'orange' too using pop() pointing to the index of it

fruits.pop(0)
# fruits is now: ["lemon"]
print(fruits)
````

````python
# You have a List of 10 countries to Travel to
countries=['South Korea', 'Turkey', 'Oman', "Japan", 'Trinidad and Tobago', 'Germany', 'Singapore', 'Norway', 'South Africa', 'Iceland']

# Print the first country

countries[0]
# Print the last country

countries[-1]
````

````python
# Find me the location of 'Germany' in the list. Using index()

countries.index('Germany')
# Print the first 5 countries

countries[0:5]

# Print the last 5 countries

countries[-5:]

# print 1 skip 1 from the countries starting from the first (using stepsize 2 in slicing)

countries[0::2]
````

````python
# Sort the list. Using listname.sort()

countries.sort()
print("Sorted list:")
print(countries)
````

````python
number = [11,17,16,7,5,6,12,18,1,13,10,8,3,7,12,20,19,21,4,9,14,7,2,15]

# Get the Minimum of the list using Min(listname)

min(number)

# # Get the Maximum of the list using Max(listname)

max(number)

# Count how many 7's are in our number list using list.count()

number.count(7)

# how many items/count of numbers in our list. Use len(listname)

len(number)
# Sum all the items in the list using sum(listname)
sum(number)

# Cauculate the Average (Mean), the sum of all items divided by the count of them
# note: there is a direct way to get the mean, it will be covered later

sum(number)/len(number)
````

### **Python Librarian**

````python
# 1.9 Python Librarian

authors = {
    "Arundhati Roy": {
        "genre": "literary fiction",
        "books": [
            "The God of Small Things",
            "The Ministry of Utmost Happiness"
        ],
        "active": True
    },
    "Brandon Sanderson": {
        "genre": "fantasy",
        "books": [
            "The Way of Kings",
            "Words of Radiance",
            "Oathbringer"
        ],
        "active": True,
        "phone": {
            "home": "(281) 330-8004",
            "work": "(877) CASH-NOW"
        }
    },
    "Kobo Abe": {
        "genre": "absurdist fiction",
        "books": [
            "The Woman in the Dunes",
            "The Face of Another"
        ],
        "active": False
    },
}
````

````python
# A. Access and print Arundhati Roy's first book

authors['Arundhati Roy']['books'][0]
````

````python
# B. We need to contact Brandon at work: access and print their work phone number

authors['Brandon Sanderson']['phone']['work']
````

````python
# C. Create a list composed of all the genres available in our library
genres = []  # this creates empty list for you. You need to use append() to add items
genres.append(authors['Arundhati Roy']['genre'])
genres.append(authors['Brandon Sanderson']['genre'])
genres.append(authors['Kobo Abe']['genre'])



print(genres)
````

````python
# D. Create a list of all books available in our library
books = []  # this creates empty list for you. You need to use append() to add items

books.append(authors['Arundhati Roy']['books'])
books.append(authors['Brandon Sanderson']['books'])
books.append(authors['Kobo Abe']['books'])


print(books)
````

````python
# E. Arundhati has decided to expand their range of genres and tackle non-fiction
#    Modify their "genre" property to now be a list that also includes non-fiction

authors['Arundhati Roy']['genre'] = ['literary fiction', 'non-fiction']
authors
````

````python
# F. Brandon Sanderson is retiring. Change their active property to False.
#    BONUS: This means we should no longer contact them. Delete their "phone" property

authors['Brandon Sanderson']['active'] = False
authors

````
