#### 1. Convert the NPAS_parsed_trunc_long.csv dataset from long to wide


```python
import pandas as pd
```

```python
# load the data

npas = pd.read_csv('data/NPAS_parsed_trunc_long.csv')
```

```python
# check the head

npas.head()

```

```python
# check shape

npas.shape
```

You can see that the long format data has far more rows than the wide data set but only three columns.

Below you can view the three columns: `subject_id`, `variable`, and `value`.

**`subject_id:`**
- This is the primary "key" or `ID` column. Each `subject_id` will have corresponding entries in the `variable` column — one for each row.

**`variable:`**
- This column indicates the variable with which the item in the `value` column corresponds.

**`value:`**

- This contains all values for all variables for all IDs. Essentially, every cell in the wide data set except the `subject_id` is listed in this column.

```python
# filter subject_id ==1

npas[npas['subject_id'] ==1]
```

**Print out the unique values in the `variable` column.**

You can see that the unique values in the `variable` column correspond to the column headers in the wide format data.

```python
# show unique values in the variable column

npas['variable'].unique()
```

 **Convert from long to wide**



```python
npas.head(1)
```

```python
# convert from long to wide

npas_wide = npas.pivot(index='subject_id',columns='variable', values='value').reset_index()

# check the head

npas_wide.head()
```

```python
# check the shape of the new created wide dataframe

npas_wide.shape

```

```python
# Remove the column label

npas_wide.columns.name = None

# check the head again

npas_wide.head()
```

```python
# save wide data to csv as nerdy_wide.csv

npas_wide.to_csv('nerdy_wide.csv', index=False)


```

#### 2. Convert it back from wide to long
- Keep `subject_id` and `major` as the `Columns that stay locked`

```python
# check the head of nerdy_widea

npas_wide.head()
```

```python
# do the convert

npas_long = pd.melt(npas_wide,id_vars=['subject_id','major'], var_name='variable', value_name='value')

# check the head

npas_long.head()
```

```python
# get the shape of the dataset

npas_long.shape
```

#### 3. Convert again from wide to long but with a selected value_vars
- Create the same DataFrame with `.melt()` on the full wide data set, but select the columns 'anxious','bookish','calm' to be used with the `value_vars` argument
- name the new dataframe with a different name than the previous
 

```python
# do the convert, call the dataframe with some other name


npas_long_new = pd.melt(npas_wide,id_vars=['subject_id','major'],value_vars= ['anxious','bookish','calm']
                                                                           ,var_name='variable', value_name='value')

# check the head

npas_long_new.head()
```

```python
# get the shape of the dataset

npas_long_new.shape
```
