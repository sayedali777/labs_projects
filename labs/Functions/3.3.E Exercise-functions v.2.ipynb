##### Q5.C
Write a function that accepts one parameter, a list
The function should return the sum total of all numbers in the list, using a loop.
\
*Hint*: Do we want a for loop or a while loop for this kind of question?

```python
# your code here
def sum_list(numlist):
    total = 0
    for num in numlist :
        total = total + num
    return total
```

```python
my_list=[3,7,1,9]

sum_list(my_list)
```

#### Q6. Problem-Solving Functions

```python
# A. Write a function that finds the longest string in a given list of strings

# your code here

def longest_string_find(stringn):
    longeststr= ''
    for s in stringn:
        if len(s) > len(longeststr):
            longeststr = s
    return longeststr
        
    

```

```python
s=["dfaf","adfaffaf","grrtsrfsgfgfsgsgf","adfaaa"]

longest_string_find(s)
```
